```python
class Test:
    a=10  # variable 1 , directly inside of class & outside of method 
    
    def __init__(self): # constructor
        Test.b=10  # variable 2, using class name
        
    def m1(self):  # instance method
        Test.c=30 # varilabe 3, using class name
        
    @classmethod
    def m2(cls):
        cls.d1=40 # variable 4, using class name
        Test.d2=444 # variable 5 , using class name
    @staticmethod
    def m3():
        Test.e=50 # variable 5, using class name
        
        
```

# Methods

* Instance method, Stastic Method, Class Method

Instance method: if we are using instance variable inside the method is called instance method, we should pass self in declaration.

```python
class Student:
    School = 'Ravi Skills'
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3
    def avg(self):
        return (self.m1+self.m2+self.m3)
s1 = Student(1,2,3)
s2 = Student(4,5,6)
```

```python
print(s1.avg())
```

```python
print(s2.avg())
```

there are two types of instance, 
1. Accessor Methods 
2. Mutator Method

if we want to fetch the values, we use accessor

if we want to change the values, we use Mutator


```python
class Student:
    School = 'Ravi Skills'
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3
    def avg(self):
        return (self.m1+self.m2+self.m3)
    def get_m1(self):
        return self.m1
    def set_m1(self,value):
        set.m1= value
        
s1 = Student(1,2,3)
s2 = Student(4,5,6)
```

```python
print(s1.get_m1())
```

```python
print(s1.set_m1(10))
```

Class method: Inside method if we use CLS variable then it is called Class method

above the method we have to declare @classmethod decorator

```python
class Student:
    School = 'Ravi Skills'
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3
        
    def avg(self):
        return (self.m1+self.m2+self.m3)
        
    def info(cls):
        return cls.school

        
s1 = Student(1,2,3)
s2 = Student(4,5,6)
```

```python
print(Student.info())
```

```python
class Student:
    School = 'Ravi Skills'
    def __init__(self,m1,m2,m3):
        self.m1=m1
        self.m2=m2
        self.m3=m3
        
    def avg(self):
        return (self.m1+self.m2+self.m3)
    @classmethod   
    def info(cls):
        return cls.School

        
s1 = Student(1,2,3)
s2 = Student(4,5,6)
```

```python
print(Student.info())
```
